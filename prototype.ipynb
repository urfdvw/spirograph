```python
import numpy as np
import matplotlib.pyplot as plt


theta = 0 # position of gear
alpha = 0 # rotation of the gear
dt = 0.01
teeth = 0.1
nout = 96
nin = 33
rout = nout * teeth / np.pi /2
rin = nin * teeth / np.pi /2
p = 0.8
plt.figure()
x_list = []
y_list = []
while (theta * rout / teeth) < np.lcm(nout, nin):
    theta = theta + dt
    alpha = alpha + dt - dt * rout / rin 
    # center of gear
    x = np.cos(theta) * (rout - rin)
    y = np.sin(theta) * (rout - rin)
    # position of hole
    x += np.cos(alpha) * rin * p
    y += np.sin(alpha) * rin * p
    x_list.append(x)
    y_list.append(y)
plt.plot(x_list, y_list)
plt.show()
```
